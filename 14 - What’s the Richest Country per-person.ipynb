### What's the richet country in the world on a per-person basis?

```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('data/countries.csv')
data.head()
```

```python
mean_gdp_per_cap=data.groupby(['country']).mean().gdpPerCapita
mean_gdp_per_cap.head()
```

```python
top5=mean_gdp_per_cap.sort_values(ascending=False).head(5)
top5.head()
```

```python
#is kuwait really the richest country int the world (in a per-person basis)?
kuait=data[data.country=='Kuwait']
kuait.head()
```

```python
plt.plot(kuait.year, kuait.gdpPerCapita)
plt.title('GDP Per Capita in Kuwait')
plt.show()
```

```python
plt.subplot(311)
plt.title('GDP Per Capita')
plt.plot(kuait.year, kuait.gdpPerCapita)

plt.subplot(312)
plt.title('GDP in Billions')
plt.plot(kuait.year, kuait.gdpPerCapita*kuait.population / 10**9)

plt.subplot(313)
plt.title('Population in millons')
plt.plot(kuait.year, kuait.population)

plt.tight_layout()
plt.show()
```

```python
# que tan rapido creció la población en Kuwait relativamente al GDP
plt.plot(kuait.year, kuait.population/kuait.population.iloc[0]*100)

kuait_gdp=kuait.population*kuait.gdpPerCapita
plt.plot(kuait.year, kuait_gdp/kuait_gdp.iloc[0]*100)
plt.plot(kuait.year, kuait.gdpPerCapita/kuait.gdpPerCapita.iloc[0]*100)

plt.title('GDP and Population Growth in Kuait (first year = 100)')
plt.legend(['Population','GDP','GDP Per Capita'])
plt.show()
```
