```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('data/countries.csv')
data.head()
```

```python
# cheack data in 2007
data_2007=data[data.year==2007]
data_2007.head()
```

```python
plt.scatter(data_2007.gdpPerCapita, data_2007.lifeExpectancy, 5) #5-> size of points
plt.title('GDP Per capita and Life Expectancy in 2007')
plt.xlabel('GDP Per Capita ($)')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
data_2007.gdpPerCapita.corr(data_2007.lifeExpectancy) #Finding correlation
```

```python
# we should see the data in a log scale
np.log10([10,100,1000])
```

```python
plt.scatter(np.log10(data_2007.gdpPerCapita),data_2007.lifeExpectancy, 5)
plt.title('Log GDP Per capita and Life Expectancy in 2007')
plt.xlabel('GDP Per Capita ($)')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
np.log10(data_2007.gdpPerCapita).corr(data_2007.lifeExpectancy)
```

```python
# check data for each year
years_sorted=sorted(set(data.year))
print(years_sorted)
```

```python
file_names=[]
for given_year in years_sorted:
    data_year=data[data.year==given_year]
    plt.scatter(data_year.gdpPerCapita, data_year.lifeExpectancy, 5)
    plt.title(given_year)
    plt.xlim(0, 60000)
    plt.ylim(25, 85)
    plt.xlabel('GDP Per Capita')
    plt.ylabel('Life Expectancy')
    #plt.show()
    name='data/'+str(given_year)
    plt.savefig(name,dpi=200) #dpi = dots  per inch
    file_names.append(name)
    plt.clf() #clear the currentplot
```

```python
import imageio
images_log = []
for filename in file_names:
    images_log.append(imageio.imread(filename+".png"))
imageio.mimsave('data/images.gif', images_log, duration=0.8)
```

```python
file_names_log=[]
for given_year in years_sorted:
    data_year=data[data.year==given_year]
    plt.scatter(np.log10(data_year.gdpPerCapita), data_year.lifeExpectancy, 5)
    plt.title(given_year)
    plt.xlim(2, 5)
    plt.ylim(25, 85)
    plt.xlabel('GDP Per Capita (Log Scale)')
    plt.ylabel('Life Expectancy')
    #plt.show()
    name_log='data/'+'log_'+str(given_year)
    plt.savefig(name_log,dpi=200) #dpi = dots  per inch
    file_names_log.append(name_log)
    plt.clf() #clear the currentplot
```

```python
import imageio
images_log = []
for filename in file_names_log:
    images_log.append(imageio.imread(filename+".png"))
imageio.mimsave('data/images_log.gif', images_log, duration=0.8)
```
