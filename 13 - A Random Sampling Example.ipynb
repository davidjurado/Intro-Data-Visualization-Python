```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('data/obama.csv', parse_dates=['year_month'])
data.head()
```

```python
data_big=pd.read_csv('data/obama_too_big.csv', parse_dates=['year_month'])
data_big.head()
```

```python
print(data.shape)
print(data_big.shape)
#rows,columns
```

```python
plt.plot(data_big.year_month, data_big.approve_percent, 'o', markersize=2, alpha=0.3)
plt.show()
```

```python
#sacamos un muestreo
sampled=data_big.sample(frac=0.1)#saca un muestreo equivalente al 10% de los datos (otra forma es n=1000)

plt.plot(sampled.year_month, sampled.approve_percent, 'o', markersize=2, alpha=0.2)
plt.show()
```
